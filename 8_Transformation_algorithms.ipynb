## Warp and swirl functions from skimage.transform

```python
import math
import matplotlib.pyplot as plt

from skimage import io, data, transform
from skimage.transform import swirl, warp
```

```python
tree = io.imread("images/tree.jpeg")

plt.figure(figsize=(8,8))
plt.imshow(tree,cmap='gray')
```

## Shape-preserving transformations

Only scaling, rotation and translation of the image

```python
tform = transform.SimilarityTransform(scale = 1.5, rotation=math.pi/4,
                                     translation=(tree.shape[0]/2,-100))
rotated = warp(tree,tform) # use warp to specify the image & the transformation you want to apply
back_rotated = warp(tree,tform.inverse)
```

```python
plt.figure(figsize=(8,8))
plt.imshow(rotated)
```

```python
plt.figure(figsize=(8,8))
plt.imshow(back_rotated)
```

Swirl

```python
tree_swirled = swirl(tree,rotation=0, strength=20, radius=150, mode = 'constant')

plt.figure(figsize=(8,8))
plt.imshow(tree_swirled)
```
