#### Corner detection using Harris corner detector

```python
from matplotlib import pyplot as plt
from skimage import data
from skimage.feature import corner_harris, corner_subpix, corner_peaks
from skimage.transform import warp, AffineTransform
from skimage.draw import ellipse, circle
```

```python
checkerboard = data.checkerboard()
plt.figure(figsize=(6,6))
plt.imshow(checkerboard,cmap="gray")
```

#### Apply transformation to the image with AffineTransform() and warp()

```python
transform = AffineTransform(scale=(0.9,0.8),rotation=1,shear=0.6,translation=(150,-80))

warped_checkerboard = warp(checkerboard,transform,output_shape=(320,320))

plt.figure(figsize=(6,6))
plt.imshow(warped_checkerboard,cmap="gray")
```

```python
corners = corner_harris(warped_checkerboard)

plt.figure(figsize=(6,6))
plt.imshow(corners, cmap="gray")
```

```python
coords_peaks = corner_peaks(corners,min_distance=1) 
# returns all of the x & y coordinates of the corners in this image
coords_peaks.shape
# there are 104 corners in this image
```

```python
# corner_subpix() performs a statistical test to determine we passed in should be an intersection of two edges, or a single peak
# window_size is a search window for subpixel estimation
coords_subpix = corner_subpix(warped_checkerboard, coords_peaks, window_size = 10)
```

```python
coords_peaks[0:11]
```

```python
coords_subpix[0:11]

# note that the corner_subpix() has moved the coordinates a bit
```

```python
fig, ax = plt.subplots(figsize=(8,8))

ax.imshow(warped_checkerboard, interpolation="nearest",cmap="gray")

ax.plot(coords_peaks[:,1],coords_peaks[:,0],".b",markersize=30)

# red stars are the ones after statistical estimation
ax.plot(coords_subpix[:,1],coords_subpix[:,0],"*r",markersize=10)

plt.tight_layout()
plt.show()
```
