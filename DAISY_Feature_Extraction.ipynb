DAISY: Generates low dimensional, invariant descriptors that can be used for image matching and classification. It is more performant and lower error than SIFT

```python
import skimage
from skimage import color, data
from skimage.transform import resize
from skimage.feature import daisy
import cv2
import matplotlib.pyplot as plt
```

```python
img1 = color.rgb2gray(skimage.io.imread("images/irish_passport.jpg"))
```

```python
plt.figure(figsize=(8,8))
plt.imshow(img1, cmap='gray')
```

```python
## extract DAISY descriptors
desc_array, desc_image = daisy(img1, 
                               step=150, # distance between the descriptor sampling points
                               radius=30, # radius of the outermost ring used for descriptor calculation
                                        # this is a circular gaussian window to calculate the histogram of gradients
                               visualize=True)
```

```python
# number of descriptors
desc_num = desc_array.shape[0] * desc_array.shape[1]
desc_num
```

```python
plt.figure(figsize=(8,8))
plt.imshow(desc_image)
```
